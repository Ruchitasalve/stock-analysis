```python
import yfinance as yf

# Download Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Reset index
tesla_data.reset_index(inplace=True)

# Display the first five rows
print(tesla_data.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL for Tesla revenue data (Yahoo Finance)
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Get the webpage content
html_data = requests.get(url).text
soup = BeautifulSoup(html_data, "html.parser")

# Find all tables
tables = soup.find_all("table")

# Check if tables were found
if not tables:
    print("No tables found on the page.")
else:
    for table in tables:
        if "Revenue" in str(table):
            tesla_revenue = pd.read_html(str(table))[0]
            break

# Ensure tesla_revenue is defined
if 'tesla_revenue' in locals():
    # Clean the data
    tesla_revenue.columns = ["Date", "Revenue"]
    tesla_revenue.dropna(inplace=True)

    # Display the last five rows
    print(tesla_revenue.tail())
else:
    print("Revenue data not found.")
```

```python
# Download GameStop stock data
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")

# Reset index
gme_data.reset_index(inplace=True)

# Display the first five rows
print(gme_data.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL for GameStop revenue data (Yahoo Finance or MacroTrends)
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"

# Get the webpage content
html_data = requests.get(url).text
soup = BeautifulSoup(html_data, "html.parser")

# Find all tables
tables = soup.find_all("table")

# Check if tables were found
if not tables:
    print("No tables found on the page.")
else:
    for table in tables:
        if "Revenue" in str(table):
            gme_revenue = pd.read_html(str(table))[0]
            break

# Ensure gme_revenue is defined
if 'gme_revenue' in locals():
    # Clean the data
    gme_revenue.columns = ["Date", "Revenue"]
    gme_revenue.dropna(inplace=True)

    # Display the last five rows
    print(gme_revenue.tail())
else:
    print("Revenue data not found.")
```

```python
import matplotlib.pyplot as plt

# Function to make the stock graph
def make_graph(data, title):
    plt.figure(figsize=(10,5))
    plt.plot(data["Date"], data["Close"], label="Stock Price", color='blue')
    plt.xlabel("Date")
    plt.ylabel("Closing Price (USD)")
    plt.title(title)
    plt.legend()
    plt.show()

# Plot Tesla stock graph
make_graph(tesla_data, "Tesla Stock Price History")
```

```python
# Plot GameStop stock graph
make_graph(gme_data, "GameStop Stock Price History")
```
